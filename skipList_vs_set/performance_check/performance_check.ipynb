- let us define closness
- closenss = (1 - (abs(T1 - T2) / max(T1, T2))) * 100

```python
import statistics

def printCloseness(skip_list_time, set_time, closeness):
    # Print table header
    print("| S.No | Skip List Time | Set Time | Closeness (%) |")
    print("|------|----------------|----------|---------------|")
    
    # Print table rows
    for i, (skip_time, set_t, close) in enumerate(zip(skip_list_time, set_time, closeness), start=1):
        print(f"| {i:<4} | {skip_time:<14.6f} | {set_t:<8.6f} | {close:<13.6f} |")



def findClosenssBetweenSkiplistAndSet():
    with open( '../execution_time/skip_list_time.txt', 'r') as skip_list_file, open('../execution_time/set_time.txt', 'r') as set_file_time:
        skip_list_time = []
        for line in skip_list_file:
            skip_list_time.append(float(line.split()[1]))

        set_time = []
        for line in set_file_time:
            set_time.append(float(line.split()[1]))

        n = len(skip_list_time)
        closeness = []
        for i in range(n):
            t1 = skip_list_time[i]
            t2 = set_time[i]
            closeness.append((1 - abs(t1 - t2)/max(t1, t2)) * 100)

        printCloseness(skip_list_time, set_time, closeness)
        averageCloseness = sum(closeness) / len(closeness)
        print("Average closeness: ", averageCloseness)
        print("Median closeness: ", statistics.median(closeness))
```
